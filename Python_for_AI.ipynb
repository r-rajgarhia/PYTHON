  3.Operators in Python

```python
#Operators
a=11
b=25
#Arithematic operators
#+,-,*,/,//,%,**
print("Division",a/b)
print("Floor Division",a//b)
print("Modulus",a%b)
print("Exponentiation",a**b)
```

```python
#Logical operators
a=True
b=False
print("a and b:",a and b)
print("a or b:",a or b)
print("not b:",not b)
```

```python
#Assignment operators
x=5
x=x*2
print("updated x:",x)
```

```python
#Identity Operators
print("Is a same as b?", a is b)
print("Is a not same as b?", a is not b)

kirana= ["salt", "sugar", "pepper", "coils", "curd", "notebook"]
print("shampoo" in kirana)
kirana.append("shampoo")
print(kirana)
print("shampoo" in kirana)

#Membership Operators
fruits = ["apple", "banana", "cherry"]
print("Is 'apple' in fruits?", "apple" in fruits)
print("Is 'mango' not in fruits?", "mango" not in fruits)
```
