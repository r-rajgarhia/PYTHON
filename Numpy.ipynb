**Array Indexing and Slicing**

**Explain how to access elements in arrays and slice them for sub-arrays.**

```python
ar1 = np.array([[[12,2,34,54,56,67],[12,23,34,45,78,89]],[[12,12,34,54,56,89],[10, 20, 30, 40, 50,90]]])
ar1[1,0,2:]
```

```python
# Indexing and slicing
arr = np.array([10, 20, 30, 40, 50])

# Accessing single element
print("First element:", arr[0])

# Slicing the array
print("Elements from index 1 to 3:", arr[1:4])

# Multi-dimensional indexing
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print("Element in first row, second column:",  matrix[0, 1])

# Slicing 2D arrays
print("Second row:", matrix[1, :])
```

**4. Basic Operations on Arrays**

**Discuss element-wise operations like addition, multiplication, and broadcasting.**

```python
arr8= np.arange(32)
print(arr8)
arr9=arr8.reshape(2,2,2,2,2)
print(arr9)
arr10 = arr9.flatten()
arr10
```

```python
# Element-wise operations
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

# Element-wise addition
print("Array addition:", arr1 + arr2)

# Element-wise multiplication
print("Array multiplication:", arr1 * arr2)

# Broadcast
arr = np.array([1, 2, 3])
print("Array after scalar multiplication:", arr * 3)
```

**Reshaping Arrays**

**Demonstrate reshaping of arrays and converting a 2D array to 1D.**

```python
# Reshaping arrays
arr = np.arange(12)
reshaped_arr = arr.reshape(3, 4)
print("Reshaped array (3x4):\n", reshaped_arr)

# Flattening the array
flattened_arr = reshaped_arr.flatten()
print("Flattened array:", flattened_arr)

# Using ravel()
ravel_arr = reshaped_arr.ravel()
print("Ravelled array:", ravel_arr)
```

**Demonstrate matrix multiplication and transpose.**

```python
# Matrix multiplication
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])
result = np.dot(matrix1, matrix2)
print("Matrix multiplication result:\n", result)
print(matrix1)
# Transpose
print("Transpose of matrix1:\n", matrix1.T)
```

**Boolean Indexing and Conditional Operations**

**Explain how to filter arrays using Boolean indexing.**

```python
# Boolean indexing
arr = np.array([10, 20, 30, 40, 50])

# Filtering based on condition
print("Elements greater than 30:", arr[arr > 30])

# Using np.where
indices = np.where(arr > 30)
print("Indices where values are greater than 30:", indices)
```

```python
marks=np.array([30,56,12,89,90])
print(marks[marks>33])
print(np.where(marks>33))
```
